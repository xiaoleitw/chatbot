```python
import csv

reader = csv.reader(open("cities.txt"), delimiter="\t")
result = []
for i in reader:
    result += i

all_cities = []

def extract_cities(province, extra):
    name = province + extra
    return list(map(lambda x: [x[len(name):-1], province, extra], list(filter(lambda x: x.startswith(name), result))))
    
provices = ['青海','甘肃','陕西','云南','贵州','四川','海南',
            '广东','湖南','湖北','河南','山东','江西','福建',
            '安徽','浙江','江苏','黑龙江','吉林','辽宁','山西','河北']

auto_prov = ['新疆','宁夏','内蒙','广西','西藏']

huge_cities = ['北京','上海','天津','重庆']

for i in huge_cities:
    all_cities += extract_cities(i, '市')

for i in provices:
    all_cities += extract_cities(i, '省')
    
for i in auto_prov:
    all_cities += extract_cities(i, '')
    
all_cities += list(map(lambda x: [x, x, '市'], huge_cities))
```

```python
with open('cities.csv', 'w', encoding='utf-8') as csvfile:
    city_writer = csv.writer(csvfile, delimiter='\t')
    for i in all_cities:
        city_writer.writerow(i)
```
